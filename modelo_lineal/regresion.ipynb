# Modelos lineales

En esta actividad se realizará un ejercicio de regresión lineal en Python 3.8.8 dentro de un documento Jupyter Notebook, usando la librería `scikit-learn` para hacer el modelo.

Primeramente importaremos las librerías necesarias:

```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn as sk
from sklearn import linear_model
from sklearn import datasets
```

Para aplicar el modelo de regresión lineal, usaremos un conjunto de datos de diabetes:

```python
diabetes_x, diabetes_y = datasets.load_diabetes(return_X_y=True)
# usar solo una columna
diabetes_x = diabetes_x[:, np.newaxis, 2]
# separar datos en entrenamiento y prueba
x_train = diabetes_x[:-20]
y_train = diabetes_y[:-20]
x_test = diabetes_x[-20:]
y_test = diabetes_y[-20:]
```

Ya que tenemos los datos, podemos proseguir a crear el modelo de regresión lineal y ajustarlo a nuestros datos:

```python
regression = linear_model.LinearRegression()
# alimentar modelo con datos
regression.fit(x_train, y_train)
# coeficientes del modelo
regression.coef_
```

Con el modelo ya entrenado, podemos hacer predicciones con los datos que separamos para pruebas:

```python
# predecir datos de prueba
y_pred = regression.predict(x_test)
# coeficiente de determinación (1 = predicciones perfectas)
sk.metrics.r2_score(y_test, y_pred)
```

Finalmente, veamos la gráfica de la regresión lineal:

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, 'black')

plt.show()
```
