# Algoritmos genéticos

Se realizará un ejercicio con código para demostrar el funcionamiento de los algoritmos genéticos, en Python 3.8 usando la librerías `geneticalgorithm` y `numpy`.

```python
import numpy as np
import geneticalgorithm as ga
```

Supongamos que queremos minimizar la función $f(X) = x_1^2 - x_2 \sqrt{x_3} + 5 \frac{x_4}{x_1}$, donde $X = \{x_1, x_2, x_3, x_4\}$ son números enteros en el rango $[1, 100]$. Este es un problema simple pero nos ayuda a entender mejor la demostración de la librería y el algoritmo genético:

```python
import math

def function(x):
    return x[0] ** 2 - (x[1] * math.sqrt(x[2])) + 5 * (x[3] / x[0])

boundaries = np.array([[1, 100]] * 4)
```

Ahora definimos los parámetros para el algoritmo:

```python
params = {
    'max_num_iteration': None,
    'max_iteration_without_improv': 100,
    # tamaño de la población
    'population_size': 100,
    # probabilidad de mutación
    'mutation_probability': 0.2,
    # probabilidad y tipo de cruce
    'crossover_probability': 0.5,
    'crossover_type': 'uniform',
    # proporción de élites
    'elit_ratio': 0.02,
    # proporción de generación anterior que se queda
    'parents_portion': 0.3,
}
```

Estos parámetros nos van a servir para configurar el modelo del algoritmo genético, al cual también le vamos a pasar la función objetivo definida y los rangos de las variables:

```python
genetic_algorithm = ga.geneticalgorithm(
    function=function,
    dimension=len(boundaries),
    variable_boundaries=boundaries,
    variable_type='int',
    algorithm_parameters=params
)
```

Finalmente, ejecutamos el algoritmo genético:

```python
genetic_algorithm.run()
```

Como era de esperarse, las variables que más hacían que la función objetivo incrementara fueron minimizadas a 1, y las que más hacían que decrementara fueron maximizadas a 100. Nuestro modelo de algoritmo genético nos devuelve la mejor solución encontrada y la función objetivo evaluada con ella.
